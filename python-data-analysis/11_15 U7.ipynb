```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.rand(100,2) #生出100筆資料 每筆都有兩個數字
```

```python
X[0]
```

```python
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
```

開K-Means機器的方法: 機器變數名=KMeans(n_cluster=你要分幾群)

```python
clf=KMeans(n_clusters=3)
```

```python
clf.fit(X)
```

```python
clf.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=clf.labels_)
```
