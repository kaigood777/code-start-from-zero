## 4-6 圖形的互動之一

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
x=np.linspace(0,2*np.pi,1000)
```

```python
def draw(n):
    y=np.sin(n*x)
    plt.plot(x,y,lw=5)
    plt.show()
```

```python
draw(3)
```

```python
interact(draw,n=(1.,50.))
```

## 4-7 圖形的互動之 over fitting

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact_manual
```

```python
X=np.random.rand(6)
Y=np.random.rand(6)
```

```python
plt.scatter(X,Y, s=100,c='r')
```

```python
x=np.linspace(0,1,1000)
```

```python
def draw(n):
    plt.scatter(X,Y, s=100,c='r')
    y=0.5*np.sin(n*x*2*np.pi)+0.5
    plt.plot(x,y)
    plt.show()
```

```python
interact_manual(draw,n=(1.,200.)) #interact_manual要按按鈕才會改變產生的的東西
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact_manual
```

```python
def BMIcal(身高, 體重):
    身高=float(身高)/100
    體重=float(體重)
    bmi=體重/(身高**2)
    print("你的BMI={:.2f}".format(bmi))
```

```python
interact_manual(BMIcal,身高="請輸入你的身高",體重="請輸入你的體重")
```
